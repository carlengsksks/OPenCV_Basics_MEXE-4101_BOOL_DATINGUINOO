## **Lesson 2: Edge Detection**

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

image = cv2.imread("IMAGES/COLORS.jpeg")
# cv2_imshow(image)

gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
canny_image = cv2.Canny(gray,150, 200)
# cv2_imshow(canny_image)

# Erosion and Dilation
kernel = np.ones((5,5), np.uint8)

#Dilation
dilate_image = cv2.dilate(canny_image, kernel, iterations=1)
# cv2_imshow(dilate_image)

#Erosion
# kernel = np.ones((1,1), np.uint8)
erode_image = cv2.erode(dilate_image,kernel, iterations=1)
# cv2_imshow(erode_image)

display = np.hstack((canny_image,dilate_image,erode_image))
cv2_imshow(display)
```

## **Lesson 4: Drawing Shapes and Writing Text on Images**

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Create a blank image
img = np.zeros((512, 512, 3), np.uint8)  # uint8: 0 to 255

# Draw the house base (Rectangle)
cv2.rectangle(img, (150, 250), (350, 450), (0, 0, 255), -1)  # Red filled rectangle

# Draw the roof (Triangle)
pts = np.array([[150, 250], [350, 250], [250, 150]], np.int32)
pts = pts.reshape((-1, 1, 2))
cv2.polylines(img, [pts], isClosed = True, color=(0, 255, 0), thickness=5)  # Green outline
cv2.fillPoly(img, [pts], color=(0, 255, 0))  # Green filled triangle

# Draw the door (Rectangle)
cv2.rectangle(img, (220, 350), (280, 450), (139, 69, 19), -1)  # Brown filled rectangle

# Draw windows (Squares)
cv2.rectangle(img, (170, 270), (200, 300), (255, 255, 0), -1)  # Yellow filled square
cv2.rectangle(img, (300, 270), (330, 300), (255, 255, 0), -1)  # Yellow filled square

# Draw a sun (Circle)
cv2.circle(img, (400, 100), 50, (0, 255, 255), -1)  # Yellow filled circle

# Draw grass (Line)
cv2.line(img, (0, 450), (512, 450), (0, 128, 0), 10)  # Thick green line

# Write a title (Text)
cv2.putText(img, "My House", (170, 50), cv2.FONT_HERSHEY_COMPLEX, 1, (255, 255, 255), 2)

# Display the image
cv2_imshow(img)
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Load the image and convert to HSV
image = cv2.imread("IMAGES/SHAPES.jpg")
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Define HSV ranges for the specified colors

# Purple
# lower_purple = np.array([130, 50, 50])
# upper_purple = np.array([160, 255, 255])

# Teal
# lower_teal = np.array([85, 50, 50])
# upper_teal = np.array([100, 255, 255])

# Green
# lower_green = np.array([45, 50, 50])
# upper_green = np.array([85, 255, 255])

# Yellow
# lower_yellow = np.array([20, 50, 50])
# upper_yellow = np.array([45, 255, 255])

# Pink
lower_pink = np.array([160, 50, 50])
upper_pink = np.array([180, 255, 255])



mask = cv2.inRange(hsv,lower_pink,upper_pink)
# cv2_imshow(mask)
result = cv2.bitwise_and(image, image, mask = mask)
cv2_imshow(result)
# cv2_imshow(image)
```

## **Lesson 2: Face Detection**

```python
import cv2
from google.colab.patches import cv2_imshow

face_cascade = cv2.CascadeClassifier("FILE/haarcascade_frontalface_default.xml")
# img = cv2.imread("images/person.jpg")
img = cv2.imread("IMAGES/GROUP.jpg")
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

faces = face_cascade.detectMultiScale(gray,1.3,5)
# print(faces)
for (x,y,w,h) in faces:
  cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),3)

cv2_imshow(img)
```
